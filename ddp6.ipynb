```python

#no.2
print('\n====== No 1 ======')
jumlah = 0 #untuk menghitung hasil jumlah
string = "" #untuk menampung nilai jumlah dalam string
bilangan = 1
while bilangan <= 19: #melalukan perulangan sebanyak <= 19
     jumlah += bilangan
     string += str(bilangan)
     if bilangan < 19:
        string += "+"
     bilangan += 2
print(f"{string} = {jumlah}")
```
